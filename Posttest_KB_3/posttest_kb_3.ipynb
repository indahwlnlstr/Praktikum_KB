```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('CarRentalData.csv')
df.head()
```

```python
df.info()
```

## PIE CHART

```python
data_pie = df['fuelType'].value_counts().rename_axis('fuelType').reset_index(name='Id_count')
plt.figure(figsize=(10, 12))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.fuelType, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('fuelType')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Pada diagram pie diatas dapat disimpulkan bahwa type terbanyak ada pada gasoline dan type yang paling sedikit terdapat pada type Diesel. Dengan persenan 1.2% diesel, 4.7% type Hybrid, 10.8% type Electic, dan 83.3% type Gasoline

## DIAGRAM BATANG

```python
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='vehicletype', palette='flare')
plt.title("Type Kendaraan")
plt.ylabel("Jumlah")
plt.xlabel("Tipe Kendaraan")
plt.grid()
plt.show()
```

pada grafik batang berikut ada 5 tipe kendaraan pada data set ini. Tipe kendaraan car lebih banyak dari pada tipe yang lainnya. 

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='vehicletype', palette='flare', hue='fuelType')
plt.title("Perbandingan Bahan Bakar Pada Jenis Kendaraan")
plt.ylabel("count")
plt.xlabel("VehicleType")
plt.show()
```

Pada grafik ini dapat dilihat bahwa perbandingan bahan bakar dan jenis kendaraan. Pada tipe van tidak terdapat bahan bakar Electric dan Hybrid dan juga pada tipe van ini jumlahnya paling sedikit dibandingkan dengan yang lainnya.
